```python
# 기본적인 그래프

import matplotlib.pyplot as plt

data = [10,15,19,30,22]

plt.plot(data);
```

```python
# 1- 그래프 틀(영역)을 만들기
# 2- 그래프 그리기
# 3- 그래프 보여주기

plt.figure(figsize(18,8))
plt.plot(data)
plt.show()
```

```python
from matplotlib import font_manager, rc
import matplotlib.pyplot as plt

plt.rcParams['axes.unicode_minus']=False  # 추가설정 : 폰트를 변경하면 -표시가 ㅁ으로 변경되기에 '-'를 변경하지 않도록 지정
rc('font', family='Malgun Gothic')
```

```python
# x축과 y축의 라벨 추가

plt.figure(figsize(18,8))
plt.plot(data)

plt.xlabel('데이터')
plt.xlabel('값')
plt.title('제목')

plt.show()
```

1- 그래프 틀(영역)을 만들기
   > plt.figure()       
    
2- 그래프 그리기
   > plt.plot(data)     
    
3- 그래프 보여주기
   > plt.show()         

 
 ### *그래프 출력시 객체표시 같은 거 안 출력하게 하려면 뒤에 세미콜론(;) 붙이면 된다*

```python
# y축 데이타에 x축 데이타를 지정

data = [10,15,19,30,22]

step = ['one', 'two', 'three', 'four','five' ]

plt.plot(step,data)

```

```python
# 속성 : linestyle , color , marker
plt.plot(step, data, color='red', linestyle='dashed', marker='o')
```

```python
# 속성 축약

plt.plot(step,data,'ro--')
```
